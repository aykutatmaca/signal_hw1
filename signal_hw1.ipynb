```python
def unit_step(a, n):
    result = []
    for sample in n:
        if sample<a:
            result.append(0)
        else:
            result.append(1)
    return np.array(result)

def unit_impulse(n):
    result = np.zeros(len(n))
    result[np.where(n == 0.0)] = 1
    return np.array(result)
```
