### Setup imports and input signals

```python
import numpy as np
import matplotlib.pyplot as plt

lower_limit = -100
upper_limit = 100

steps_discrete = 1
steps_continous = 0.01

x_discrete = np.arange(lower_limit, upper_limit + steps_discrete, steps_discrete)
x_continous = np.arange(lower_limit, upper_limit + steps_continous, steps_continous)
```

## Question 1

### Part A

```python
def unit_step(a, x):
    result = []
    for sample in x:
        if sample<a:
            result.append(0)
        else:
            result.append(1)
    return np.array(result)

def unit_impulse(x):
    result = np.zeros_like(x)
    result[np.where(x == 0.0)] = 1
    return np.array(result)
```

```python
y_1 = 5 * np.exp(5 * x_discrete)
y_2 = unit_step(0, x_discrete) + unit_step(-50, x_discrete) + unit_impulse(x_discrete)
y_3 = -1 * unit_step(-3, x_continous) + 2 * unit_step(-1, x_continous) + -2 * unit_step(1, x_continous) + unit_step(3, x_continous)
y_4 = np.sin(x_continous * np.pi) + np.cos(np.pi / 4 * x_continous)

plt.figure(figsize=(15, 30))

plt.subplot(4,1,1) #not periodic
plt.title("y 1")
plt.stem(x_discrete, y_1)

plt.subplot(4,1,2) #not periodic
plt.title("y 2")
plt.stem(x_discrete, y_2)

plt.subplot(4,1,3) #not periodic
plt.title("y 3")
plt.plot(x_continous, y_3)

plt.subplot(4,1,4)  #PERIODIC Period = 8
plt.title("y 4")
plt.plot(x_continous, y_4)

plt.show()

#NOT PERIODIC
```
